```python
%%writefile naukan.twol
Alphabet
p m f v ɢ w t n n̥ s ʃ l̥ l r ɳ j̊ j k ŋ q x ɣ ɴ χ a e i o u ɨ
;

Sets
Consonant =p m f v w t n n̥ s ʃ l̥ l r ɳ j̊ j k ŋ q x ɣ ɴ χ ɢ;
Sonorant = l l̥ m n̥ n ŋ r;
Vowel = a e i o u ɨ;
Long = ː;

Rules
"Space"
%{Space%}:%  <=> _ ;
"n to t"
n:t <=> Consonant|Vowel _ Consonant|Vowel;
"mn to n"
mn:n <=> Consonant|Vowel _ Consonant|Vowel;
"ɢ to χ"
ɢ:χ <=> Consonant|Vowel _ t;
"ia to eː"
ia:eː<=> Consonant _ Consonant
;
```
